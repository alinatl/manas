```python
def insert_lemma(token):
    req="""SELECT idlemma FROM lemma WHERE txt='{}'""".format(token.token['lemma'])
    lemmas = cur.execute(req)
    if lemmas != None:
        lemma = lemmas['idlemma']
    else:
        req="""INSERT INTO lemma (txt) VALUES ('{}')""".format(token.token['lemma'])
        cur.execute(req)
        req = 'SELECT LAST_INSERT_ID()'
        lemma = cur.execute(req)
    return lemma

def get_POS(token):
    req="""SELECT idmorphtag FROM morphtag WHERE tag='{}'""".format(token.token['upos'])
    cur.execute(req)
    req = 'SELECT LAST_INSERT_ID()'
    POS = cur.execute(req)['idmorphtag']
    return POS

def insert_sent(start, end):
    req="""INSERT INTO sentence (id_start_sent, id_end_sent)
    VALUES ({},{})""".format(start, end)
    cur.execute(req)
    req = 'SELECT LAST_INSERT_ID()'
    sent = cur.execute(req)['idsentence']
    return sent

def get_edgelabel(token):
    req="""SELECT idedgelabel FROM edgelabel WHERE txt='{}'""".format(token.token['deprel'])
    cur.execute(req)
    req = 'SELECT LAST_INSERT_ID()'
    edgelabel = cur.execute(req)['idedgelabel']
    return edgelabel


def insert_parent(id_Parentlabel, id_parent, id_edgelabel, id_sentence):
    req="""INSERT INTO parent (id_Parentlabel, id_parent, id_edgelabel, id_sentence)
    VALUES ({},{})""".format(id_Parentlabel, id_parent, id_edgelabel, id_sentence)
    cur.execute(req)
    req = 'SELECT LAST_INSERT_ID()'
    return parent
```
